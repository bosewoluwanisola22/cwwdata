```python
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
durations = [103,101,99,100,100,95,95,96,93,90]
movie_dict = {
    'years' : [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020],
    'durations' : [103,101,99,100,100,95,95,96,93,90]
}
print(movie_dict)
```

```python
import pandas as pd
duration_df = pd.DataFrame.from_dict(movie_dict)
print(duration_df)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
duration_df.plot(x = 'years', y = 'durations', kind = 'line')
plt.title("A simple line plot of years against durations")
plt.show()
```

```python
netflix_df = pd.read_csv('netflix_data.csv')
print(netflix_df.head())
```

```python
netflix_df_only = netflix_df[(netflix_df["type"] == "Movie")]
netflix_movies_col_subset = netflix_df[["title", "country", "genre", "release_year", "duration"]]
print(netflix_df_only.head())
```

```python
fig = plt.figure(figsize=(12,8))
netflix_movies_col_subset.plot(x = 'release_year', y = 'duration', kind = 'scatter')
plt.title("A simple scatter plot of release_year against duration")
plt.show()
```

```python
short_movies = netflix_movies_col_subset[(netflix_movies_col_subset["duration"] < 60)]
print(short_movies.head(20))
```

```python
colors = []
for genre, row in netflix_movies_col_subset.iterrows() :
    if row['genre'] == "Children" :
        colors.append('red')
    elif row['genre'] == "Documentaries" :
        colors.append('blue')
    elif row['genre'] == "stand_up" :
        colors.append('green')
    else:
        colors.append('black')
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colors)
plt.title('Movie duration by year of release')
plt.xlabel('Release year')
plt.ylabel('Duration(min)')
```
